# Seleção do Modelo de Regressão - PyCaret

```python
import pandas as pd
from pycaret.regression import *
```

```python
dados = pd.read_csv('dados_anuncios_airbnb_tratados.csv')
```

```python
dados.drop(columns='Título', inplace=True)
```

```python
dados.head()
```

## Setup Ambiente

```python
regression_setup = setup(data = dados,
                         target = 'Preço (R$)',
                         numeric_features=['Avaliação', 'Hóspedes', 'Camas', 'Quartos', 'Banheiros'],
                         categorical_features=['Bairro', 'Imóvel', 'Cozinha', 'Wi-Fi', 'Piscina'],
                         session_id=123)
```

## Melhor modelo de regressão

```python
best = compare_models()
```
