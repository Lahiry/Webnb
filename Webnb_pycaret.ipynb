# Seleção do Modelo de Regressão - PyCaret

```python
import pandas as pd
from pycaret.regression import *
```

```python
dataset = pd.read_csv('dados_anuncios_airbnb_tratados.csv')
```

```python
dataset.drop(columns='Título', inplace=True)
dataset = dataset.dropna()
dataset.head()
```

```python
data = dataset.sample(frac=0.7, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

## Setup Ambiente

```python
regression_setup = setup(data = data,
                         target = 'Preço (R$)',
                         numeric_features=['Avaliação', 'Hóspedes', 'Camas', 'Quartos', 'Banheiros'],
                         categorical_features=['Bairro', 'Imóvel', 'Cozinha', 'Wi-Fi', 'Piscina'],
                         session_id=0)
```

## Melhor modelo de regressão

```python
best = compare_models(sort='RMSE')
```
